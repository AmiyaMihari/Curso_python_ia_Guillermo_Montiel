# Clase de Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Bases

```python
x = [1,2,3,4,5]
y = [10,20,22,27,35]

#Graficar
plt.scatter(x,y)
```

```python
#Cambiar el color
plt.plot(x, y, color='red')
```

```python
#Cambiar el color
plt.plot(x, y, color='red', marker='o')
```

```python
# Cambiar el color
plt.plot(x, y, color='red', marker='o', linestyle='--')

# Titulos
plt.title('Mi primer grafico')

# Etiquetas de los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Hacer malla (cuadrícula)
plt.grid(True)

# Mostrar la grafica
plt.show()
```

# Ejes

```python
plt.xlim(0, 5)
plt.ylim(0, 10)
```

```python
plt.xticks([1, 2, 3, 4], ["uno", "dos", "tres", "cuatro"])
plt.yticks([1, 2, 3, 4], ["uno", "dos", "tres", "cuatro"])
```

```python
plt.grid(True, linestyle='--', alpha=1)
plt.show()
```

```python
ax = plt.gca() # Obtenemos los ejes
ax.spines['top'].set_color('red') # Cambiamos el color de la parte superior
ax.spines['top'].set_linewidth(2) # Cambiamos el grosor de la parte superior
ax.spines['right'].set_color('red') # Cambiamos el color de la parte derecha
# Desaparecemos la parte izquierda
ax.spines['left'].set_color('none')
```

# Tipos de gráficos

### Dispersión (Scatter)

```python
horas = [1,2,3,4,5]
calificaciones = [60,70,80,75,90]

plt.scatter(horas, calificaciones,alpha=0.6, s=100, c='#5a9', marker='^')
plt.title('Horas de estudio vs calificaciones')
```

### Barras

```python
# Gráficos de barras
panes = ["Pan de muerto", "Roca de reyes", "Bigotes"]
ventas = [50, 45, 30]

plt.bar(panes, ventas, color='red', alpha=0.6, edgecolor='black', linewidth=1.5)
```

### Pastel

```python
# Gráficas de pastel
etiquetas = ["Pan de muerto", "Roca de reyes", "Bigotes", "Galleta", "Mantecadas"]
valores = [50, 42, 35, 20, 30]
explotar = (0.2, 0, 0, 0, 0) # Separa las rebanadas
colores = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']

plt.pie(valores, labels=etiquetas, explode=explotar, colors=colores)
plt.show()
```
