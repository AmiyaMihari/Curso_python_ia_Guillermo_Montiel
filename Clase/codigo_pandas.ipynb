```python
customers.to_csv("customers.csv", index=False)
```

# Selección de datos

```python
# Seleccionamos las compañias q aparecen
companias = customers["Company"]
print(companias)
```

```python
# Elegir únicamente aquellas cuyo index tienen valor mayor a 50
indices = customers[customers["Index"] > 50]
gente_cuyo_nombre_empieza_con_a = customers[customers["First Name"].str.startswith("a")]
suscritos_en_2020 = customers[customers["Subscription Date"].str.contains("2020")]

# iloc: Acceso por posición
primer_cliente = customers.iloc[0]
print(primer_cliente)
```

# Agregar datos

```python
# Agregar columnas
prebes = pd.DataFrame({
    "Nombre": ["Pepito", "Leo", "Deivi", "Bere", "Javi"],
    "Edad": [40, 21, 26, 26, 5],
    "Nota": [10, 0, 0, 10, 0],
})

#Agregarme a mí mismo
prebes.loc[5] = ["Jair", 26, 10]

# Agregar Columna de Sexo
prebes["Sexo"] = ["M", "M", "M", None, "M", "M"]

prebes
```

# Operaciones con datos

```python
# Promedio
prebes["Nota"].mean()
float(prebes["Nota"].mean()) # Promedio de notas en float
```

# F-Strings
Son una forma más fácil de formatear strings en Python 

```python
media_nota = float(prebes["Nota"].mean())
print(f"Los prebes obtuvieron {media_nota} en promedio de sus notas")

prebes["Edad en 50 años"] = prebes["Edad"] + 50
print(prebes)
```
